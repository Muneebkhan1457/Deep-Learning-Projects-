```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.txt',sep = ';',header = None,names = ['text','emotion'])
```

```python
df.isnull().sum()
```

```python
df['text']
```

```python
df['emotion'].unique()
```

```python
df['emotion'] = df['emotion'].map({'sadness':0,'anger':1,'love':2,'surprise':3,'fear':4,'joy':5})
```

```python
df
```

```python
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)
df['text'] = df['text'].apply(remove)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['emotion'], test_size=0.2, random_state=42
)

# Initialize CountVectorizer
vect = CountVectorizer()

# Fit on training data ONLY
X_train_bow = vect.fit_transform(X_train)

# Transform test data using the SAME vocabulary
X_test_bow = vect.transform(X_test)

# Train MultinomialNB
nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)

# Predict on test set
pred_bow = nb_model.predict(X_test_bow)

# Evaluate
print("Accuracy:", accuracy_score(y_test, pred_bow))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Initialize TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Fit on training data
X_train_tfidf = tfidf.fit_transform(X_train)

# Transform test data using the SAME vocabulary
X_test_tfidf = tfidf.transform(X_test)

# Train Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)

# Predict
pred_tfidf = nb_model.predict(X_test_tfidf)

# Evaluate
print("Accuracy:", accuracy_score(y_test, pred_tfidf))
```
