```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,stratify=y)
```

```python
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(80, activation='relu', input_shape=(X_train.shape[1],)),
    BatchNormalization(),
    Dropout(0.2),

    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.1),

    Dense(16, activation='relu'),
    Dropout(0.1),

    Dense(1, activation='sigmoid')  # Output for binary classification
])
```

```python
from tensorflow.keras.optimizers import Adam

# Define optimizer with custom learning rate
optimizer = Adam(learning_rate=0.0005)

# Compile the model
model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=15,           # increase patience slightly
    restore_best_weights=True
)
```

```python
model.fit(X_train,y_train,validation_split=0.2,epochs=100,batch_size=32,verbose=1,callbacks=[early_stop])
```

```python
loss,acc=model.evaluate(X_train,y_train)
print(acc)
```

```python
loss,acc=model.evaluate(X_test,y_test)
print(acc)
```
